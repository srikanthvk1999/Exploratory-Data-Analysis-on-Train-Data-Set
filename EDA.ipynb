EDA : Exploratory Data Analysis - Understanding what is there inside the data

converting unstructured data into structured data

Iteration 01 :
1. Read to the Data and print head
2. Shape of the data
3. Information about the data
4. Finding the duplicate rows, drop the duplicate rows
5. observing statistical analysis of the continuous and categorical
6. Rename the columns (optional)
7. imputation of meanmedian to continous data and mode imputer to categorical data
8. fit the data
9. transform the data
10. final data with no null values

Iteration 02:

1. Univariate Analysis

continous=hist,box,kde,LMplot

categorical=count,piechart

2. BI-variate Analysis

continous,categorical=hist,box,kde plot

categorical,categorical=countplot,piechart

continous,continous=scatter,Lmplot

3. Multi-variate Analysis

scatter plot,
pair plot,Heat map

```python
data_encoded=data.copy()
```
